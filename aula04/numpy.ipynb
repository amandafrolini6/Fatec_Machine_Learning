### 2. Criando Arrays(Vetores e Matrizes)

A principal estrutura de dados do NumPy é o ndarray

```python
# Criando um array 1D (Vetor) a partir de uma lista Python
vetor = np.array([1, 2, 3, 4, 5])
print("Vetor 1D:", vetor)

# Criando um array 2D (Matriz) a partir de uma lista de listas
matriz = np.array([[1, 2, 3], [4, 5, 6]])
print(f"\n {matriz[0, 0]}")
print("\nMatriz 2D:\n", matriz)
```

### Funções úteis para criação de Arrays

O NumPy possui funções nativas para gerar arrays comuns rapidamente, o que é muito útil para inicializar pesos em Redes Neurais, por exemplo.

```python
# Array de zeros
zeros = np.zeros((2,3)) # 2 linhas e 3 colunas
print(f"Zeros:\n {zeros}")
```

```python
# Array de uns
uns = np.ones((3, 3))
print(f"\nUns:\n {uns}")
```

```python
# Array com uma sequência de números (similar ao range() do Python)
sequencia = np.arange(0, 10, 2) # Início, Fim (exclusivo), Passo
print(f"\nSequência (arange): {sequencia}")
```

```python
# Valores linearmente espaçados
espacados = np.linspace(0, 1, 5) # 5 valores entre 0 e 1
print(f"\nLinspace: {espacados}")
```

### 3. Inspecionando Arrays

É crucial saber o tamanho e o formato dos seus dados, especialmente ao preparar datasets para mineração de dados.

```python
print(f"Formato (shape): {matriz.shape}") # (linhas, colunas)
print(f"Dimensões (ndim): {matriz.ndim}") # Número de dimensões
print(f"Tamanho total (size): {matriz.size}") # Total de elementos
print(f"Tipo de dado (dtype): {matriz.dtype}") # Tipos dos elementos (ex: int64, int32)
```

### 4. Operações Matemáticas Básicas

O NumPy permite fazer operações elemento a elemento de forma direta, sem a necessidade de loops for (conceito conhecido como vetorização).

```python
a = np.array([1, 2, 3])
b = np.array([10, 20, 30])

print(f"Soma: {a + b}")
print(f"Subtração: {b - a}")
print(f"Multiplicação: {a * b}")
print(f"Divisão: {b / a}")
```

```python
# Funções estatísticas básicas
dados = np.array([15, 20, 25, 30, 35])
print(f"Média: {np.mean(dados)}")
print(f"Desvio Padrão: {np.std(dados)}")
print(f"Valor Máximo: {np.max(dados)}")
```

### 5. Indexação e Fatiamento (Slicing)

Você pode acessar elementos específicos ou extrair subconjuntos do array.

```python
arr = np.array([10, 20, 30, 40, 50])

print(f"Primeiro elemento: {arr[0]}")
print(f"Último elemento: {arr[-1]}")
print(f"Fatiamento (índice 1 ao 3): {arr[1:4]}") # O índice final é exclusivo
```

```python
# Em matrizes 2D
matriz_ex = np.array([[10, 20, 30], [40, 50, 60], [70, 80, 90]])
print(f"\nMatriz completa:\n {matriz_ex}")
print(f"\nElemento na linha 1, coluna 2: {matriz_ex[1,2]}")
print(f"\nTodas as linhas, apenas coluna 1:\n {matriz_ex[:, 1]}")
```
