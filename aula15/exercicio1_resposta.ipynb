EXERCÍCIO 1 - QUESITOS

**a) Quais variáveis mais influenciaram a decisão de promoção?**

R: As variáveis experiência, quantidade de treinamentos e idade apresentaram influência equilibrada na decisão de promoção.

Os testes foram realizados utilizando combinações das variáveis de 2 em 2:

idade × experiência;

idade × treinamentos;

experiência × treinamentos.

Todas as combinações obtiveram 100% de acurácia no conjunto de treino. Dessa forma, não foi possível identificar uma única variável como significativamente mais importante que as demais.

Os gráficos mostraram que todas as variáveis conseguiram separar bem os funcionários promovidos dos não promovidos, indicando que todas contribuíram de maneira relevante para a classificação.

**b) A árvore consegue separar bem os funcionários promovidos dos não promovidos?**

R: Sim. A árvore de decisão conseguiu separar relativamente bem os funcionários promovidos dos não promovidos.

Isso pode ser observado: pela acurácia elevada obtida no treinamento, pelas regiões bem definidas nos gráficos de fronteira de decisão e pela divisão clara entre as classes.

**c) Há padrões claros, como por exemplo “funcionários com X anos de experiência e Y treinamentos tendem a ser promovidos”?**

R: Sim. Foram observados alguns padrões claros no dataset.

Exemplos:

Funcionários com maior experiência tendem a possuir maior chance de promoção;

Funcionários com mais treinamentos realizados também apresentam maior probabilidade de serem promovidos;

Em muitos casos, funcionários com mais experiência e maior número de treinamentos ficaram concentrados na região classificada como “promovido”.

Isso mostra que a árvore conseguiu aprender relações entre as variáveis e a decisão final de promoção.

**d) Como a profundidade da árvore impacta a capacidade de generalização do modelo nesse dataset?**

R: A profundidade da árvore influencia diretamente a capacidade de generalização do modelo.

Árvores muito rasas: podem simplificar demais o problema, não conseguem capturar padrões importantes e podem causar underfitting.

Árvores muito profundas: aprendem detalhes excessivos do conjunto de treino, podem memorizar os dados e aumentam o risco de overfitting.

Nesse exercício, utilizar max_profundidade = 3 ajudou a equilibrar: boa separação das classes e sem tornar o modelo excessivamente complexo.

Assim, o modelo conseguiu aprender padrões importantes sem perder muita capacidade de generalização.
