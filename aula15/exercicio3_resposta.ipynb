EXERCÍCIO 3 - QUESITOS

**a) Quais fatores foram mais relevantes para prever o churn: meses de contrato, valor da fatura ou uso do suporte técnico?**

R: Os fatores meses de contrato, valor da fatura e uso do suporte técnico contribuíram para a previsão de churn, porém o fator meses de contrato apresentou maior destaque.

Isso foi observado nos testes realizados com combinações de variáveis de 2 em 2:

meses × suporte técnico;

meses × valor da fatura;

valor da fatura × suporte técnico.

As combinações que incluíam a variável meses de contrato alcançaram melhor desempenho, chegando a 100% de acurácia no conjunto de treino. Isso indica que essa variável teve maior importância na separação entre clientes que permaneceram e clientes que cancelaram o serviço.

Os gráficos também mostraram que o tempo de contrato ajudou bastante a definir regiões mais claras de classificação do churn.

**b) Há algum perfil de cliente que o modelo identifica como mais propenso a cancelar?**

R: Sim. Pelos gráficos gerados pela árvore de decisão, foi possível identificar alguns padrões associados ao cancelamento.

Clientes com menos meses de contrato, maior valor de fatura ou maior utilização do suporte técnico apresentaram maior tendência ao churn em determinadas regiões dos gráficos.

Isso indica que clientes mais recentes ou que apresentam mais problemas e necessidade de suporte podem possuir maior probabilidade de cancelar o serviço.

**c) O modelo apresenta mais facilidade em identificar clientes que permanecem ou que cancelam?**

R: O modelo apresentou bom desempenho na identificação das duas classes, já que obteve alta acurácia no conjunto de treino.

Pelos gráficos, foi possível observar regiões relativamente bem separadas entre clientes que permanecem e clientes que cancelam. Isso mostra que a árvore conseguiu aprender padrões importantes do dataset.

**d) Como a empresa poderia usar os insights da árvore de decisão para reduzir a taxa de churn?**

R: A empresa poderia utilizar os resultados da árvore de decisão para identificar clientes com maior risco de cancelamento e agir preventivamente.

Por exemplo: oferecer promoções para clientes mais novos e melhorar o atendimento do suporte técnico.

Dessa forma, a empresa poderia aumentar a satisfação dos clientes e reduzir a taxa de cancelamento.
