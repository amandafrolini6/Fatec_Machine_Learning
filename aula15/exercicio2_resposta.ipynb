EXERCÍCIO 2 - QUESITOS

**a) Quais critérios foram mais determinantes para a aprovação de crédito: renda, idade ou score de crédito?**

R: Os critérios renda, idade e score de crédito apresentaram influência equilibrada na aprovação de crédito.

Os testes foram realizados utilizando combinações das variáveis de 2 em 2:

renda × idade;

idade × score;

renda × score.

Todas as combinações obtiveram 100% de acurácia no conjunto de treino. Dessa forma, não foi possível identificar uma única variável como mais importante que as demais.

Os gráficos mostraram que todas as variáveis contribuíram de maneira relevante para separar os clientes com crédito aprovado dos clientes com crédito negado.

**b) Existe um limiar de score ou de renda a partir do qual o crédito quase sempre é aprovado?**

R: Sim. Pelos gráficos gerados pela árvore de decisão, foi possível observar regiões onde clientes com maior renda e maior score de crédito tendem a ter aprovação de crédito com maior frequência.

Renda + Idade = 4000 e 30 anos
Idade + Score = 30 anos e 55 score
Renda + Score = 3000 e 55 score

Isso indica que existem determinados valores de score e renda que aumentam significativamente a chance de aprovação.

**c) O modelo apresentou erros de classificação? Quais poderiam ser as consequências disso para o banco?**

R: Não. O modelo apresentou 100% de acurácia no conjunto de treino, indicando que não houve erros de classificação nos dados utilizados para treinamento.

Mesmo assim, em situações reais, erros podem acontecer. Caso o modelo aprove crédito para clientes com alto risco de inadimplência, o banco pode sofrer prejuízos financeiros. Por outro lado, negar crédito para clientes confiáveis pode causar perda de clientes e redução de oportunidades de negócio.

**d) A árvore é fácil de interpretar para justificar ao cliente por que o crédito foi aprovado ou negado?**

R: Sim. A árvore de decisão é considerada um modelo fácil de interpretar, pois as decisões são feitas com base em regras simples, como renda, idade e score de crédito.

Isso facilita explicar ao cliente os motivos da aprovação ou negação do crédito, mostrando quais critérios foram utilizados na decisão. Essa transparência é uma das principais vantagens da árvore de decisão.
