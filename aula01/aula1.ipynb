```python
import matplotlib.pyplot as plt
anos = [2020, 2021, 2022, 2023, 2024]
vendas = [100, 120, 150, 130, 160]

plt.plot(anos, vendas)
plt.title('Vendas ao Longo dos Anos')
plt.xlabel('Anos')
plt.ylabel('Vendas (em milhões)')
plt.grid(True)
plt.show()
```
