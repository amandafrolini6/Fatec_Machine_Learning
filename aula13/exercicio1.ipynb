```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("dataset_exercicio_1.csv", "r") as f:
    linhas = f.readlines()

dados = []
for linha in linhas:
    linha = linha.strip().strip('|')
    if '---' in linha:
        continue
    partes = [p.strip() for p in linha.split('|')]
    try:
        dados.append([float(p) for p in partes])
    except ValueError:
        continue

dataset = np.array(dados)

print(dataset)
```

```python
X = dataset[:, 0]
y = dataset[:, 1]
n = len(X)

print("X:", X)
print("y:", y)
print("\n Número de amostras: ", X.shape[0])
```

```python
x_mean = np.mean(X)
y_mean = np.mean(y)

print("Média de X:", x_mean)
print("Média de y:", y_mean)

b, a = np.polyfit(X, y, 1)
print(f"\nIntercepto (a): {a:.4f}")
print(f"Inclinação (b): {b:.4f}")

y_pred = a + b * X
print(f"\nValores previstos: {y_pred}")

coef_pearson = np.corrcoef(X, y)[0, 1]
print(f"\nCálculo do Coeficiente de Pearson: {coef_pearson:.4f}")

det_quadrado = coef_pearson ** 2
print(f"\nCálculo de Determinação: {det_quadrado * 100:.4f}")

plt.scatter(X, y, label="Dados")
plt.plot(X, y_pred, color="red", label="Regressão")
plt.title("Regressão Linear", fontsize=14)
plt.xlabel("Investimento em Marketing(mil R$)", fontsize=12)
plt.ylabel("Faturamento(mil R$)", fontsize=12)
plt.text(0.05, 0.9, f'$R^2 = {det_quadrado:4f}$', transform=plt.gca().transAxes, fontsize=14, verticalalignment='top', bbox=dict(boxstyle='round,pad=0.5', fc='wheat', alpha=0.5))
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```


### a- O modelo de regressão linear conseguiu representar bem a relação entre investimento em marketing e faturamento? Justifique observando o gráfico.

R: Sim, o modelo de regressão linear conseguiu representar bem a relação, pois o gráfico mostra uma tendência crescente clara. À medida que o investimento em marketing aumenta, o faturamento também cresce, indicando uma relação aproximadamente linear entre as variáveis.

### b- Qual é o coeficiente angular (inclinação da reta) e como ele pode ser interpretado no contexto do problema?

R: A inclinação da reta é aproximadamente b ≈ 3,6. Isso significa que, para cada aumento de Ré 3,6 mil.

### c- Se a empresa investir R$ 15 mil em marketing, qual seria o faturamento previsto pelo modelo?

Assim, o faturamento previsto seria de aproximadamente R$ 67,9 mil.

### d- Existem indícios de que o crescimento do faturamento pode ser limitado (isto é, que a relação pode não ser linear para investimentos muito altos)?

R: Pelo gráfico apresentado, a relação parece linear dentro do intervalo analisado. No entanto, isso não garante que o crescimento continuará linear para valores muito altos de investimento, já que, na prática, podem existir limitações como saturação de mercado ou retorno decrescente.

### e- Você considera que apenas o marketing explica o faturamento da empresa ou outros fatores deveriam ser incluídos em um modelo mais realista?

Outros fatores, como preço dos produtos, qualidade, concorrência, reputação da empresa e condições econômicas, também influenciam significativamente e deveriam ser considerados em um modelo mais realista.
