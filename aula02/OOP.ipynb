## Fundamentos de Python para Aprendizagem de Máquina

Este notebook revisita os conceitos base necessários para qualquer desenvolvimento em Python, focando em sintaxe e estruturas de dados.

## 1. Variáveis e Tipos de Dados

Python é dinamicamente tipado, o que facilita a declaração de variáveis.

```python
inteiro = 10
ponto_flutuante = 20.5
texto = "Python no Fatec"
booleano = True

print(f"Tipos: {type(inteiro)}, {type(ponto_flutuante)}, {type(texto)}, {type(booleano)}")
```


## 2. Estruturas de Dados (Listas, Tuplas e Dicionários)

Essas são as estruturas mais utilizadas para organizar informações

```python
# Lista (Mutável)
frutas = ["maçã", "banana", "cereja"]
frutas.append("uva")

# Tupla (Imutável)
coordenadas = (10, "Vini")

# Dicionário (Chave-Valor)
aluno = {"nome": "Alissa", "curso": "Tecnologia", "nota": 9.5}

print(f"Lista: {frutas}")
print(f"Dicionário: {aluno['nome']} tirou {aluno['nota']}")
```

## 3. Estruturas de Controle e Repetição

Controle do fluxo de execução do programa.

```python
# Condicional
media = 7.5
if media >= 6:
    resultado = "Aprovado"
else:
    resultado = "Exame"

# Loop For
numeros = [1, 2, 3, 4, 5]
quadrados = [x**2 for x in numeros]

print(f"Resultado: {resultado}")
print(f"Quadrados: {quadrados}")

print("\nLista de Numeros (jeito longo)")
quadrados = []
for x in numeros:
    quadrados.append(x**2)
print(quadrados)
```

## 4. Funções e Lambda

Modularização de código e funções anônimas.

```python
def calcular_bonus(salario, percentual=10):
    return salario + (salario * percentual / 100)

# Função Lambda (Anônima)
dobro = lambda x: x * 2

print(f"Bônus: {calcular_bonus(3000)}")
print(f"Dobro de 5: {dobro(5)}")
```
